```python
from IPython.display import display, HTML

import pandas as pd
import matplotlib.pyplot as plt

results=pd.DataFrame({"Number of trees":[5,20,50,100,200,250,500],
    "Mean Squared Error":[1104.0058405819577,982.0588355211296,959.7172013821779,950.9538818382658,948.1251201548552,946.7676813406915,946.2883722372982],
    "R Squared Score":[0.49588687566021783,0.5515705355335097,0.5617722125307132,0.5657737351970612,0.567065409423608,0.5676852455662749,0.5679041085475692]})

plt.figure(figsize=(10, 6))
plt.plot(results["Number of trees"], results["Mean Squared Error"], marker='o', color='blue', label="Mean Squared Error")
plt.title("Number of Trees vs Mean Squared Error")
plt.xlabel("Number of Trees")
plt.ylabel("Mean Squared Error")
plt.grid(True)
plt.legend()
plt.savefig("hotelmseplot.jpg")
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(results["Number of trees"], results["R Squared Score"], marker='o', color='green', label="R Squared Score")
plt.title("Number of Trees vs R Squared Score")
plt.xlabel("Number of Trees")
plt.ylabel("R Squared Score")
plt.grid(True)
plt.legend()
plt.savefig("hotelr2plot.jpg")
plt.show()
```
